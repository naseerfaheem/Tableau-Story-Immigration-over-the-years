# Clean

- Row (index 2) should be column names. 

### Define
- Choose Row 3 (index 2) as the new columns using df.loc function: 

### Code

```python
df.columns = df.loc[2]
```

### Test

```python
df.head()
```

- The first two rows and the 3rd row need to be dropped. 

### Define
- Drop the first two rows that contain null values and also the repeated rows using panda's drop function and the index of the row. 

### Code

```python
#Use the drop funtion and a range of the row indices: 
df = df.drop([0,1,2])
```

```python
#Reset the indeces
df = df.reset_index(drop=True)
```

### Test

```python
df.head()
```

```python
df.info()
```

- Tota pending start-year to Total Pending Start-Year
- statistics.filter.decisions_recognized to Recognized
- statistics.filter.decisions_other to other
- Origin to country_of_origin, Country / territory of asylum/residence to country_of_asylum, Tota pending start-year to Total_Pending_Start_Year

### Define
- rename the above column names to make them more meaningful using panda's rename function. 

```python
df.info()
```

### Code

```python
#Since we are just renaming column name, we do not have to use inplace=True becuase it will return Nonetype. 
df = df.rename(columns={'Country / territory of asylum/residence':'Country of Asylum/Residence',
                        'Origin':'Country of Origin','Tota pending start-year':'Total Pending Start_Year',
                        'statistics.filter.decisions_recognized':'Recognized',
                       'statistics.filter.decisions_other':'Other',
                       'Total pending end-year':'Total Pending End_Year'})
```

```python
#Store the cleaned file in a new csv file
df.to_csv('unhcr_asylum.csv', index=False)
```
